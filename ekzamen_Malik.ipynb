```python
def replacer(target_str, new_values):
    for i, j in new_values.items():
        target_str = target_str.replace(i, j)
    return target_str

new_values = {".": ":)",
",": ":)",
"!": ":)",
"?": ":)",
"-": ":)",
":": ":)",
";": ":)",
"—": ":)",}
string = "Тяжёлая интернет-зависимость — это когда ты выходишь из интернета, а он из тебя нет."

string = replacer(string, new_values)
print(string)
```

```python
months=[31,29,30,31,30,31,30,31,30,31,30,31]
def ordinalDate(d,m,y):
    return d+(sum(months[0:m-1]))
```
