```python
# без цикла
word = input("Word: ")
palindrom = word[::-1]
if word == palindrom:
    print("Слово палиндром")
else:
    print("Слово ne палиндром")

# цикл 
word = input("Word: ")
word_len = int(len(word))
word_len_cupl = word_len // 2
word_len = word_len - 1
l = []
l.extend(word)
for i in range(word_len_cupl):
    l[word_len - i], l[i] = l[i], l[word_len - i]
palindrom = "".join(l)
if word == palindrom:
    print("Слово палиндром")
else:
    print("Слово ne палиндром")
```
