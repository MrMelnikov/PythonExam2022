1 — если на вход поступила строка "Менеджер"
2 — в любом другом случае.</b>
</div>

```python
def profession_code(s):
    if s == 'Рабочий':
        return 0
    if s == 'Менеджер':
        return 1
    else:
        return 2
```

<div class="alert alert-block alert-warning">
<b> 11. Напишите функцию age_category, которая на вход получает число, а на выход отдаёт:
"молодой" — если возраст меньше 23
"средний" — если возраст от 23 до 35
"зрелый" — если возраст больше 35.</b>
</div>

```python
def age_category(n):
    if n < 23:
        return 'молодой'
    if 23 < n < 35:
        return 'средний'
    if n > 35:
        return 'зрелый'
```

<div class="alert alert-block alert-warning">

</div>

```python
sample['Age_category']=sample.Age.apply(age_category)
sample
```
